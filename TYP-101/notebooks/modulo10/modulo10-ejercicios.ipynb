# üß© Ejercicio 1 ‚Äî Exportaci√≥n Named

Crea un archivo `math.ts` que exporte **dos funciones**:

- `sumar(a,b)`
- `multiplicar(a,b)`

Luego imp√≥rtalas desde el notebook y pru√©balas.

### üìÑ C√≥digo a pegar en `math.ts`:

```ts
export function sumar(a: number, b: number) {
  return a + b;
}

export function multiplicar(a: number, b: number) {
  return a * b;
}
```

````typescript
// Creamos el archivo math.ts
await Deno.mkdir("modulo10/ejercicios", { recursive: true });

await Deno.writeTextFile(
  "modulo10/ejercicios/math.ts",
  `export function sumar(a: number, b: number) {
     return a + b;
   }

   export function multiplicar(a: number, b: number) {
     return a * b;
   }`
);
````

````typescript
// Importamos y usamos el m√≥dulo
import { sumar, multiplicar } from "./modulo10/ejercicios/math.ts";

console.log("sumar(2,3) =", sumar(2, 3));
console.log("multiplicar(4,5) =", multiplicar(4, 5));
````

# üß© Ejercicio 2 ‚Äî Exportaci√≥n Default

Crea un archivo `saludos.ts` que exporte **default** una funci√≥n `saludar()`.

### üìÑ C√≥digo a pegar en `saludos.ts`:

```ts

```

````typescript
// Creamos saludos.ts
await Deno.writeTextFile(
  "modulo10/ejercicios/saludos.ts",
  `export default function saludar(nombre: string) {
     return \`Hola, ${nombre}!\`;
   }`
);
````

# üß© Ejercicio 3 ‚Äî Mezcla de named + default

Crea `strings.ts` con:

- `export function capitalizar(texto)`
- `export default function longitud(texto)`

### üìÑ C√≥digo para `strings.ts`:

```ts
export function capitalizar(texto: string) {
  return texto.charAt(0).toUpperCase() + texto.slice(1);
}

export default function longitud(texto: string) {
  return texto.length;
}
```

````typescript
// Creamos strings.ts
await Deno.writeTextFile(
  "modulo10/ejercicios/strings.ts",
  `export function capitalizar(texto: string) {
     return texto.charAt(0).toUpperCase() + texto.slice(1);
   }

   export default function longitud(texto: string) {
     return texto.length;
   }`
);
````

````typescript
import longitud, { capitalizar } from "./modulo10/ejercicios/strings.ts";

console.log(capitalizar("hola"));
console.log("longitud:", longitud("TypeScript"));
````

# üß© Ejercicio 4 ‚Äî Barrel File (`index.ts`)

Crea un `index.ts` que reexporte *todos* los m√≥dulos creados hasta ahora.

### üìÑ C√≥digo para `index.ts`:

```ts
export * from "./math.ts";
export * from "./strings.ts";
export { default as saludar } from "./saludos.ts";
```

````typescript
// Creamos index.ts
await Deno.writeTextFile(
  "modulo10/ejercicios/index.ts",
  `export * from './math.ts';
   export * from './strings.ts';
   export { default as saludar } from './saludos.ts';`
);
````

````typescript
// Importamos TODO desde el barrel
import * as M from "./modulo10/ejercicios/index.ts";

console.log(M.sumar(1, 2));
console.log(M.capitalizar("typescript"));
console.log(M.saludar("Ana"));
````
